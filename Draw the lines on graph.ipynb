```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)

y1 = 2 * x + 1
y2 = 2 * x + 2
y3 = 2 * x + 3

plt.figure(figsize=(8, 6))
plt.plot(x, y1, color='blue', linestyle='-', label=r"$y = 2x + 1$")
plt.plot(x, y2, color='green', linestyle='--', label=r"$y = 2x +  2$")
plt.plot(x, y3, color='black', linestyle='-.', label=r"$y = 2x +  3$")

plt.title("Graph of lines y = 2x + 1, y = 2x + 2, y = 2x + 3")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()

plt.grid(True)

plt.show()
```
